```python
import seaborn as sns
import matplotlib.pyplot as plt

fontsize=18
file_path = './diff_init_analysis_network/diff_init_analysis_network.csv'
df = pd.read_csv(file_path)

# Define the columns of interest
columns_of_interest = ['cai', 'coop_degree', 'order_degree']

# Clean the data by removing rows with NaN or infinite values in the columns of interest
df_clean = df.replace([np.inf, -np.inf], np.nan).dropna(subset=columns_of_interest)

# Create a column that combines influence_type and interaction_type
df_clean['type_pair'] = df_clean['influence_type'] + ', ' + df_clean['interaction_type']


# Create a figure with three subplots: one for each of CAI, Coop Degree, and Order Degree with reduced height
fig, axes = plt.subplots(1, 3, figsize=(16, 5), sharey=True)

# Plot boxplot for CAI
sns.boxplot(data=df_clean, y='type_pair', x='cai', ax=axes[0])
axes[0].set_title('Cumulative Average Inclination', fontsize=fontsize)
axes[0].set_xlabel('', fontsize=fontsize)
axes[0].set_ylabel('Influence Type, Interaction Type', fontsize=fontsize)
axes[0].tick_params(axis='both', which='major', labelsize=fontsize)

# Plot boxplot for Coop Degree
sns.boxplot(data=df_clean, y='type_pair', x='coop_degree', ax=axes[1])
axes[1].set_title('Cooperativeness Degree', fontsize=fontsize)
axes[1].set_xlabel(' ')
axes[1].tick_params(axis='x', which='major', labelsize=fontsize)

# Plot boxplot for Order Degree
sns.boxplot(data=df_clean, y='type_pair', x='order_degree', ax=axes[2])
axes[2].set_title(' Order Degree', fontsize=fontsize)
axes[2].set_xlabel(' ', fontsize=fontsize)
axes[2].tick_params(axis='x', which='major', labelsize=fontsize)

plt.suptitle('')
sns.despine()

plt.tight_layout()
plt.savefig('diff_init_analysis_network.pdf')
plt.show()
```
