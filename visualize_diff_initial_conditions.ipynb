```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
fontsize=18
# Load the CSV file
file_path = './diff_initial_conditions/diff_initial_conditions.csv'
df = pd.read_csv(file_path)

# Calculate mean values for the specified columns
mean_values = df[['cai', 'coop_degree', 'order_degree']].mean()

# Adjusting the y-ticks with multiline labels for metric names
metric_names_multiline = ['CAI', 'Cooperativeness\nDegree', 'Order Degree']

plt.figure(figsize=(9, 5))

# Create a color palette for the metrics
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']

# Create violin plots with specific colors
sns.violinplot(data=df[['cai', 'coop_degree', 'order_degree']], inner='quartile', palette=colors, orient='h')

# Overlay mean points with larger and bold markers in white
for i, (col, color) in enumerate(zip(['cai', 'coop_degree', 'order_degree'], colors)):
    plt.scatter([mean_values[col]], [i], color='white', edgecolor='black', label='Mean' if i == 0 else "", zorder=10, s=150, linewidths=2)

# Update the y-ticks with multiline metric names
plt.yticks(range(len(metric_names_multiline)), metric_names_multiline, fontsize=fontsize)
plt.xticks(fontsize=fontsize)
plt.title('Multiple Runs', fontsize=fontsize)
plt.ylabel(' ')
plt.xlabel(' ')
plt.xlim(0.1, 0.95)

#plt.legend(frameon=False, fontsize=fontsize)
plt.grid(False)
sns.despine()
plt.tight_layout()
plt.savefig('diff_multiple_runs.pdf')
plt.show()
```
